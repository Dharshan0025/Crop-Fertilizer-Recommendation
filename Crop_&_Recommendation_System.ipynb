```python
features = crop.columns.to_list()
features.remove('label')
print(features)
```

```python

# Feature distributions
features = ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']
plt.figure(figsize=(15, 10))
for i, feature in enumerate(features):
    plt.subplot(3, 3, i + 1)
    sns.histplot(crop[feature], kde=True, color='blue', bins=20)
    plt.title(f"Distribution of {feature}")
plt.tight_layout()
plt.show()
```

```python
# Ploting scatter plot

plt.figure(figsize=(8, 6))
sns.scatterplot(x=crop['N'], y=crop['rainfall'], hue=crop['label'], palette="viridis", alpha=0.7)
plt.xlabel("Nitrogen (N)")
plt.ylabel("Rainfall (mm)")
plt.title("Scatter Plot: Nitrogen vs. Rainfall for Different Crops")
plt.legend(title="Crop Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Ploting scatter plot 2

import seaborn as sns
fig, ax = plt.subplots(3, 3, figsize=(15, 15))
for i, subplot in zip(features, ax.flatten()):
    sns.scatterplot(x=i, y='label', data=crop, ax=subplot)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=crop.iloc[:, :-1], palette="Set2")  # Excluding 'label' column
plt.xticks(rotation=45)
plt.title("Box Plot of Features to Detect Outliers")
plt.show()
```
